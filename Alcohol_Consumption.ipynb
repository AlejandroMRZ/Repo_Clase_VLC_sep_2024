# Ex - GroupBy

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"

pd.read_csv(url)
```

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv(url)
drinks.head(10)
```

### Step 4. Which continent drinks more beer on average?

```python
drinks["continent"].fillna('NA',inplace=True)
drinks.head(10)
```

```python
continent_beer_avg = drinks.groupby('continent')["beer_servings"].mean()
continent_beer_avg
```

### Step 5. For each continent print the statistics for wine consumption.

```python
continent_wine_consumption = drinks.groupby("continent")["spirit_servings"].describe()
continent_wine_consumption
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
mean_alcohol_consumption = drinks.groupby("continent")["total_litres_of_pure_alcohol"].mean()
mean_alcohol_consumption
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
median_alcohol_consumption = drinks.groupby("continent")[["beer_servings","spirit_servings","wine_servings"]].median()
median_alcohol_consumption
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
statistic_values_continent = drinks.groupby("continent").describe()
statistic_values_continent
```
