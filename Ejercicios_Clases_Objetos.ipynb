```python
class Tienda:
    """
    Esta clase representa una tienda de electrodomésticos.

    Atributos:
        tipo (str): El tipo de tienda. En este caso, es de electrodomésticos.
        abierta (bool): Indica si la tienda está abierta.
        nombre (str): El nombre de la tienda.
        direccion (str): La dirección de la tienda.
        numero_de_empleados (int): El número de empleados de la tienda.
        ventas_3_ultimos_meses (list): Lista de las ventas de los últimos 3 meses.
    """
    tipo = "electrodomesticos"
    abierta = True

    def __init__(self, nombre, direccion, numero_de_empleados, ventas_3_ultimos_meses):
        """
        Inicializa una nueva tienda con los valores dados.

        Args:
            nombre (str): El nombre de la tienda.
            direccion (str): La dirección de la tienda.
            numero_de_empleados (int): El número de empleados.
            ventas_3_ultimos_meses (list): Las ventas de los últimos tres meses.
        """
        self.nombre = nombre
        self.direccion = direccion
        self.numero_de_empleados = numero_de_empleados
        self.ventas_3_ultimos_meses = ventas_3_ultimos_meses

    def venta_media(self):
        """
        Calcula la venta promedio por empleado en función de las ventas
        de los últimos 3 meses.

        Returns:
            float: La venta media por empleado. Si no hay empleados, devuelve 0.
        """
        total_ventas = sum(self.ventas_3_ultimos_meses)
        if self.numero_de_empleados > 0:
            return total_ventas / self.numero_de_empleados
        return 0

    def ventas_ultimo_mes(self):
        """
        Devuelve las ventas del último mes.

        Returns:
            float: Las ventas del último mes.
        """
        return self.ventas_3_ultimos_meses[-1]

    def proyeccion_ventas_con_marketing(self, inversion_marketing):
        """
        Proyecta las ventas de los últimos meses con una inversión en marketing.

        Args:
            inversion_marketing (float): Cantidad invertida en marketing.

        Returns:
            list: Las nuevas ventas proyectadas con el impacto de marketing.
        """
        if inversion_marketing < 1000:
            factor = 1.2
        else:
            factor = 1.5

        # Aplica el factor de proyección de marketing a cada mes
        ventas_proyectadas = [venta * factor for venta in self.ventas_3_ultimos_meses]
        return ventas_proyectadas

    def nombre_and_direccion(self):
        """
        Imprime el nombre y la dirección de la tienda.
        """
        print(self.nombre, self.direccion)

# Creación de tres instancias de Tienda
tienda1 = Tienda("HogarTech", "Calle 1", 8, [5000, 2000, 1500])
tienda2 = Tienda("ElectroX", "Calle 2", 1, [5200, 5000, 6200])
tienda3 = Tienda("Tien21", "Avenida del Mar", 12, [1500, 12000, 3500])

# Acceder a atributos y métodos de la tienda 1
print(tienda1.nombre)  # HogarTech
print(tienda1.direccion)  # Calle 1
print(tienda1.numero_de_empleados)  # 8
print(tienda1.venta_media())  # Venta media por empleado
print(tienda1.ventas_ultimo_mes())  # Ventas del último mes
print(tienda1.proyeccion_ventas_con_marketing(500))  # Proyección de ventas con marketing

# Calcular las ventas del último mes de todas las tiendas
for tienda in [tienda1, tienda2, tienda3]:
    print(f"Ventas último mes de {tienda.nombre}: {tienda.ventas_ultimo_mes()}")

# Imprimir nombres de las tiendas cuya dirección contiene "Avenida"
for tienda in [tienda1, tienda2, tienda3]:
    if "Avenida" in tienda.direccion:
        print(f"Tienda en Avenida: {tienda.nombre}")
```

```python
class perro:
    """
    Clase que representa a un perro con ciertas características predeterminadas, como el número de patas, orejas y ojos,
    así como métodos que permiten al perro andar, parar y ladrar.

    Atributos de clase:
    -------------------
    - patas (int): Número de patas del perro (por defecto 4).
    - orejas (int): Número de orejas del perro (por defecto 2).
    - ojos (int): Número de ojos del perro (por defecto 2).
    - velocidad (int): Velocidad actual del perro (inicia en 0).

    Atributos de instancia:
    -----------------------
    - raza (str): Raza del perro.
    - pelo (str): Color del pelo del perro (por defecto 'marrón').
    - owner (bool): Indica si el perro tiene dueño (por defecto False).

    Métodos:
    --------
    - __init__(self, raza, pelo="marron", owner=False):
        Constructor que inicializa la raza del perro, el color de su pelo y si tiene o no dueño.

    - andar(self, aumento_velocidad):
        Método que incrementa la velocidad del perro en función del valor pasado como argumento.

    - parar(self):
        Método que detiene al perro, estableciendo su velocidad en 0.

    - ladrar(self, sonido):
        Método que hace que el perro ladre. Aunque recibe un argumento, siempre devuelve "GUAU!".
    """
    
    # Atributos de clase
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0

    def __init__(self, raza, pelo="marron", owner=False):
        """
        Inicializa una instancia de la clase perro con la raza, color de pelo, y estado de dueño.
        
        Parámetros:
        -----------
        - raza (str): La raza del perro.
        - pelo (str, opcional): El color del pelo del perro. Valor predeterminado 'marrón'.
        - owner (bool, opcional): Indica si el perro tiene dueño. Valor predeterminado False.
        """
        self.raza = raza
        self.pelo = pelo
        self.owner = owner

    def andar(self, aumento_velocidad):
        """
        Incrementa la velocidad del perro en función del valor proporcionado.

        Parámetros:
        -----------
        - aumento_velocidad (int): Valor que aumenta la velocidad del perro.
        """
        self.velocidad += aumento_velocidad

    def parar(self):
        """
        Detiene al perro, estableciendo la velocidad en 0.
        """
        self.velocidad = 0

    def ladrar(self, sonido):
        """
        Hace que el perro ladre.

        Parámetros:
        -----------
        - sonido (str): Representa el sonido del ladrido (aunque no se usa).
        
        Retorna:
        --------
        - str: El sonido del ladrido ("GUAU!").
        """
        return "GUAU!"


# Crear una instancia de la clase perro
mi_perro = perro('chihuahua')

# Comprobación de atributos
print(mi_perro.raza)       # chihuahua
print(mi_perro.ojos)       # 2
print(mi_perro.pelo)       # marrón
print(mi_perro.owner)      # False
print(mi_perro.ladrar("")) # GUAU!

# Prueba que ande, y comprueba su velocidad
mi_perro.andar(10)
print(mi_perro.velocidad)  # 10

# Páralo
mi_perro.parar()
print(mi_perro.velocidad)  # 0

```
